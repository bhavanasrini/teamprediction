```python
from keras.layers import Dense, Dropout, LSTM, Embedding,Conv1D,GlobalMaxPooling1D,Activation
from keras.preprocessing.sequence import pad_sequences
from keras.models import Sequential
import pandas as pd
import numpy as np
from numpy import array
import pickle
input_folder = '/Users/vijaybhavana/Desktop/KC_games/Data_pickle/'
train = ['KC_1stWeek.pkl','KC_2ndWeek.pkl','KC_3rdWeek.pkl','KC_4thWeek.pkl','KC_5thWeek.pkl','KC_6thWeek.pkl']
test = ['KC_7thWeek.pkl']
def load_data(test_split = 0.2):
    X_train = []
    Y_train = []
    X_test = []
    Y_test = []
    print ('Loading data...')
    for i in train:
        file = open(input_folder+i,'rb')
        f = pickle.load(file)
        X_train.append(f[0])
        Y_train.append(f[1])
        file.close()
    for i in test:
        file = open(input_folder+i,'rb')
        f = pickle.load(file)
        X_test.append(f[0])
        Y_test.append(f[1])
        file.close()
    X_train = array(X_train)
    X_test = array(X_test)
    Y_train = array(Y_train)
    Y_test = array(Y_test)
    print(X_train.shape,Y_train.shape,X_test.shape,Y_test.shape)
    return X_train, Y_train, X_test, Y_test


def create_model(input_length):
    print ('Creating model...')
    model = Sequential()
    model.add(Conv1D(filters=1, kernel_size=1, padding='valid',input_shape= (15,3)))
    model.add(GlobalMaxPooling1D())
    model.add(Dense(10, activation='sigmoid'))
    model.add(Dropout(0.5))
    model.add(Activation('relu'))
    model.add(Dense(1, activation='sigmoid'))
    print ('Compiling...')
    model.compile(loss='binary_crossentropy',
                  optimizer='rmsprop',
                  metrics=['accuracy'])
    print(model.summary())
    return model


X_train, Y_train, X_test, Y_test = load_data()

model = create_model(len(X_train[0]))

print ('Fitting model...')

hist = model.fit(X_train, Y_train, batch_size=1, epochs=100, validation_split = 0.1, verbose = 1)
print(model.predict(X_test,batch_size=1))
```
